From the above we can see that the number of singularities for a class $(a_0,a_1)$ is:
$$\begin{align*}
    &\gcd(-a_0+4a_1,2a_0-3a_1) + \gcd(-2a_0+a_1,-a_0+3a_1) \\
    &= \gcd(-a_0+4a_1,5a_1) + \gcd(-5a_1,-a_0+3a_1) \\
    &= \gcd(a_0+a_1,5a_1) + \gcd(a_0+2a_1,5a_1)
\end{align*}$$

The minimzing direction is (1,0) so the class on each level set of the Thurston norm which is (Euclidean) closest to the minimizer is simply the one with the smallest (absolute value) slope.

```python
for n in range(100):
    print('Thurston norm ' + str(n))
    smallest_slope = Infinity
    store = [0]
    for a in range(1,n+1):
        b = (2*a-n)/2
        slope = b/a
        if cone.interior_contains(a,b) and (gcd(a,b) == 1) and (slope < smallest_slope):
            smallest_slope = slope
            store[0] = (a,b)
    if store[0] != 0:
        print('Thurston norm ' + str(n) + ' has class ' + str(store[0]) + ' closest to the minimizer which has genus ' + str(get_genus(sig,store[0])))
```

```python
def get_classes_with_fixed_Tnorm_and_genus(Tnorm, genus):
    retval = []
    for a in range(1,Tnorm+1):
        b = (2*a - Tnorm)/2
        if cone.interior_contains((a,b)) and gcd(a,b) == 1 and get_genus(sig,(a,b)) == genus:
            retval.append((a,b))
            
    return retval
```

```python
for g in range(20):
    two_g_classes = get_classes_with_fixed_Tnorm_and_genus(2*g,g)
    two_g_plus_4_classes = get_classes_with_fixed_Tnorm_and_genus(2*g+4,g)
    
    min_slope_two_g = Infinity
    closest_two_g = [0]
    for term in two_g_classes:
        if abs(term[1]/term[0]) < min_slope_two_g:
            min_slope_two_g = abs(term[1]/term[0])
            closest_two_g[0] = term
    
    min_slope_two_g_plus_4 = Infinity
    closest_two_g_plus_4 = [0]
    for term in two_g_plus_4_classes:
        if abs(term[1]/term[0]) < min_slope_two_g_plus_4:
            min_slope_two_g_plus_4 = abs(term[1]/term[0])
            closest_two_g_plus_4[0] = term
            
    print('Genus: ' + str(g) + '\t Closest class with Tnorm 2g: ' + str(closest_two_g[0]) + '\t Closest class with Tnorm 2g+4:  ' + str(closest_two_g_plus_4[0]))
```

```python
def get_prim_int_lattice_pts_on_levset_arr_by_genus(Tnorm):
    # This function gets all the primitive integral points in the interior of the cone
    # on a given level set of the Thurston norm.
    preretval = []
    for a in range(Tnorm+1):
        b = (2*a - Tnorm)/2
        if cone.interior_contains(a,b) and gcd(a,b) == 1:
            preretval.append(((a,b),get_genus(sig,(a,b))))
                    
    genuses = [elt[1] for elt in preretval]
    max_genus = max(genuses)
    
    retval = [[] for i in range(max_genus + 1)]
    for elt in preretval:
        retval[elt[1]].append(elt[0])
        
    return retval
```

```python
def make_point_and_one_tnorm_plot(Tnorm):
    min_dir_arrow = arrow((0, 0), (min_dir[0]*10, min_dir[1]*10), color='cyan')
    picture = min_dir_arrow
    
    pts = get_prim_int_lattice_pts_on_levset_arr_by_genus(Tnorm)
            
    for i in range(len(pts)):
        #picture += points(pts[i], rgbcolor = rainbow(len(pts))[i], size = 50, legend_label='genus ' + str(i), label = i)
        for j in range(len(pts[i])):
            picture += text(str(i), pts[i][j], color = rainbow(len(pts))[i])
    
    Tnorm_levset = line([(2*(Tnorm/2),Tnorm/2),(0,-(Tnorm/2))])
    picture += Tnorm_levset
        
    return picture
```

```python
make_point_and_one_tnorm_plot(114)
```

```python
def get_prim_integer_lattice_points_in_cone_arranged_by_genus(maxTnorm):
    # This does the same as the previous function, but it does so for all level sets
    # up to maxTnorm.
    preretval = []
    for n in range(maxTnorm):
        for a in range(n+1):
            b = (2*a-n)/2
            if cone.interior_contains(a,b) and (gcd(a,b) == 1):
                preretval.append(((a,b),get_genus(sig,(a,b))))
                
    genuses = [elt[1] for elt in preretval]
    max_genus = max(genuses)
    
    retval = [[] for i in range(max_genus + 1)]
    for elt in preretval:
        retval[elt[1]].append(elt[0])
    
    return retval
```

```python
def make_point_and_tnorm_plot(maxTnorm):
    min_dir_arrow = arrow((0, 0), (min_dir[0]*25, min_dir[1]*25), color='cyan')
    picture = min_dir_arrow
    
        
    pts = get_prim_integer_lattice_points_in_cone_arranged_by_genus(maxTnorm)
            
    for i in range(len(pts)):
        picture += points(pts[i], rgbcolor = rainbow(len(pts))[i], pointsize = 60, legend_label='genus ' + str(i))
        
    lines = [line([(2*i,i),(0,-i)]) for i in range(floor(maxTnorm/2)+1)]
    for elt in lines:
        picture += elt
        
    return picture
```
