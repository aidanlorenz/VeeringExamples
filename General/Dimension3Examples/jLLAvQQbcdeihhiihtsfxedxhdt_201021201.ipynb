Consider the classes $(1,g,g)$ for $g$ ranging from $1$ to infinity. Using the above information,
$$ \| (1,g,g) \|_T = 1 + 2*g $$
and by the previous block, we have
$$\#\text{punctures}(1,g,g) = \gcd(g+2, -g+1) + \gcd(2g-3, -g) + \gcd(1, -1) = \gcd(3, g-1) + \gcd(3, g) + 1.$$
Thus if $g \cong 2 \pmod{3}$, then $\gcd(3, g-1) = \gcd(3,g) = 1$ whereby the number of punctures is equal to $3$. 

Thus, for $g \cong 2 \pmod{3}$,
\begin{align*}
\text{genus}(1,g,g) &= \frac{1}{2}\left( \color{red}{\| (1,g,g) \|_T} \color{blue}{- \#\text{punctures}(1,g,g)} \color{green}{+ 2}\right) \\
&= \frac{1}{2}\left(\color{red}{1+2g} \color{blue}{- 3} \color{green}{+ 2}\right) \\
&= g
\end{align*}

What is more, one can see again by the previous code block that for all such $g$, the slope on the second (using zero-indexing) boundary component is $-1/1$; in other words, it doesn't depend on $g$ so they all lie on the same slope of the second boundary component. That is, all these classes lie in the Dehn filling of the second boundary component by slope $-1/1$.

To see that these classes actually represent pseudo-Anosov monodromies after Dehn filling, we should observe that they have no one-prongs. To do this, we will use a version of the VeeringGeneralFunction get_num_prongs(sig, point), but we have to be careful because this function won't let us use the variable $g$ that we have been using. So we work through that function piece by piece being careful with the variable.

```python
point = (1,var('g'),var('g'))
substitution_dict = {get_hom_vars(sig)[i]:point[i] for i in range(len(point))} # substitute the given point for our variable place holders.
two_chain_with_free_vars = get_2_chain_from_point_in_homology(sig).subs(substitution_dict)

two_chain_with_free_vars
```

```python
no_free_vars = [] # two_chain_with_vars still contains the free variables. We want to set those all to 0. But NOT g!
for term in two_chain_with_free_vars:
    no_free_vars.append(term.subs({v:0 for v in term.variables() if v != var('g')})) # don't replace the variable g with 0 here!
    
no_free_vars = vector(no_free_vars)
no_free_vars
```

```python
total_prongs_each_boundary_comp = get_prongs_matrix(sig)*no_free_vars
total_prongs_each_boundary_comp
```

The above tells us that all punctures coming from the zeroth boundary component split $2g+1$ prongs, those coming from the first split $2g$ prongs, while those coming from the second boundary component split $1$ prong respectively. That's not good because that means the Dehn fillings aren't pseudo-Anosov.
