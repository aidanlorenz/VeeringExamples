Consider the classes $(3,g,1)$ with $g$ ranging from $1$ to infinity. Using the above, we have
$$ \| (3,g,1) \| = 3 + 2g + 2 = 5 + 2g $$
and 
$$ \# \text{punctures}(3,g,1) = \gcd(2g-1, g) + \gcd(-g-4, -g-3) + \gcd(-9,-3) = \gcd(-1,g) + \gcd(-1, -g-3) + 3 = \\ =1 + \gcd(-1,-g) + 3 = 1 + 1 + 3 = 5. $$

Hence
$$ \text{genus}(3,g,1) = \frac{1}{2}\left( \| (3,g,1) \| - \#\text{punctures}(3,g,1) + 2\right) = \frac{1}{2}\left(5 + 2g - 5 + 2\right) = g+1 $$

What is more, all such classes have boundary slope $-3/-9 = 1/3$ on the second boundary component (using zero-indexing).
So $(3,g,1) \in S_2(1/3)$.

Now we check that they have no $1$-prongs.
